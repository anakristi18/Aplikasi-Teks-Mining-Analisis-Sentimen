```python
def create_topics_dataframe(data_text, mgp, threshold, topic_dict, token_list):
    result = pd.DataFrame(columns=["tweet", "topik", "token_list"])
    for i, text in enumerate(data_text):
        result.at[i, 'tweet'] = text
        result.at[i, 'token_list'] = token_list[i]
        prob = mgp.choose_best_label(token_list[i])
        # print(prob)
        if prob[1] >= threshold:
          result.at[i, 'topik'] = topic_dict[prob[0]]
        else:
          result.at[i, 'topik'] = 'Other'
    return result
```

```python
import gensim
dictionary = gensim.corpora.Dictionary(datafinal)
```

```python
#create BOW dictionary
bow_corpus = [dictionary.doc2bow(doc) for doc in datafinal]
```

```python
from gensim.models import CoherenceModel
```

```python
def get_topics_list(model, top_clusters, n_words):
  topics = []
  for cluster in top_clusters:
    sorted_dict = sorted(model.cluster_word_distribution[cluster].items(), key=lambda k:k[1], reverse=True)[:n_words]
    topic = []
    for k,v in sorted_dict:
      topic.append(k)
    topics.append(topic)
  return topics
```

```python
#get topics to feed to coherence model
topics = get_topics_list(mgp, top_index, 10)
```

```python
#evaluate model using Topic Coherence score
cm_gsdmm = CoherenceModel(topics=topics,dictionary=dictionary,corpus=bow_corpus,texts=datafinal,coherence='c_v')
```

```python
#get coherence value
coherence_gsdmm = cm_gsdmm.get_coherence()
print(coherence_gsdmm)
```
